```python
# Truth-level selection cut functions
def get_index_truth(h5f, pdg, primary:bool):
    # Get the truth-level index of particles
    # Set primary to True for selecting only primary particle
    indices = []
    for i in range(len(h5f)):
        # IMPORTANT: In the AmBe source data there's two start process:subprocess, 2:22 and 1:91, 2:22 should be scintillation and 1:91 is unclear
        # Also the end process is always 4:131, not sure why the discrepancy
        if h5f['primary'][i]==primary and h5f['pdg_id'][i]==pdg:
            indices.append(i)
    return indices

def get_mutiplicity_truth(h5f, prim_indices):
    # Calculate how many daughter particles are generated per event
    multi = []
    for i in prim_indices:
        n = 0
        for j in range(len(h5f)):
            if h5f['parent_id'][j]==h5f['traj_id'][i] and h5f['event_id'][j]==h5f['event_id'][i]:
                n+=1
        multi.append(n)
    return multi

def get_energy_truth(h5f, indices):
    # Get energy of a segment in truth trajectories
    energy = []
    for i in indices:
        energy.append(h5f['mc_truth/trajectories/data']['E_start'][i]-h5f['mc_truth/trajectories/data']['E_end'][i])
    return energy

def event_energy_truth(h5f, evt):
    # Get the total energy of all truth level energy deposition of an event
    return sum(h5f['mc_truth/trajectories/data']['E_start'][h5f['mc_truth/trajectories/data']['event_id'] == evt]-h5f['mc_truth/trajectories/data']['E_end'][h5f['mc_truth/trajectories/data']['event_id'] == evt])

def primevent_energy_truth(h5f, evt):
    # Get the total energy of all truth level energy deposition from the primary particles of an event
    tot_E = 0
    for i in h5f['mc_truth/trajectories/data'][h5f['mc_truth/trajectories/data']['event_id'] == evt]:
        if i['parent_id']==-1:
            tot_E += i['E_start']-i['E_end']
        else:
            pass
    return tot_E

def particle_energy_truth(h5f, evt, pdg, isprim:bool):
    # Get the total energy of all truth level energy deposition from the primary particles of an event
    tot_E = 0
    for i in h5f['mc_truth/trajectories/data'][h5f['mc_truth/trajectories/data']['event_id'] == evt]:
        if isprim:
            if i['parent_id']==-1 and i['pdg_id']==pdg:
                tot_E += i['E_start']-i['E_end']
            else:
                pass
        else:
            if i['parent_id']!=-1 and i['pdg_id']==pdg:
                tot_E += i['E_start']-i['E_end']
            else:
                pass
    return tot_E
```

## AmBe Source Analysis

```python
# Generate all hits grouped by events

all_hits_bt = []     # List of all hits group by events
hits_E_truth = []    # List of truth-level energy corresponding with each event with hits
gamma_E_truth = []   # List of truth-level energy of gammas per event

# Also check what kind of processes are producing hits
hitspdg = []
hits_startprocess = []
hits_startsubprocess = []
hits_endprocess = []
hits_endsubprocess = []

for file in glob.iglob('../AmBe_Prod_V2/FLOW/*.hdf5'):
    #truth_N_index = []
    with h5py.File(file, 'r') as h5file:
        # Group truth trajectories into events
        evt_id = list(set(h5file['mc_truth/trajectories/data']['event_id']))
        for i in evt_id:
            # Find all events that have corresponding hits
            temphits = alltruth2charge(h5file, [i], 'calib_prompt_hits')
            if len(temphits)>0:
                all_hits_bt.append(temphits)
                hits_E_truth.append(particle_energy_truth(h5file, i, 22, False))
                hitspdg.append([h5file['mc_truth/trajectories/data']['pdg_id'][h5file['mc_truth/trajectories/data']['event_id'] == i]])
                hits_startprocess.append([h5file['mc_truth/trajectories/data']['start_process'][h5file['mc_truth/trajectories/data']['event_id'] == i]])
                hits_startsubprocess.append([h5file['mc_truth/trajectories/data']['start_subprocess'][h5file['mc_truth/trajectories/data']['event_id'] == i]])
                hits_endprocess.append([h5file['mc_truth/trajectories/data']['end_process'][h5file['mc_truth/trajectories/data']['event_id'] == i]])
                hits_endsubprocess.append([h5file['mc_truth/trajectories/data']['end_subprocess'][h5file['mc_truth/trajectories/data']['event_id'] == i]])
            else:
                pass
```
